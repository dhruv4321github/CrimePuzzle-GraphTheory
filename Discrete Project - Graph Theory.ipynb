```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.DiGraph()
G.add_edges_from([('A','B'),('A','C'),('A','D'),('A','E'),('A','F'),('A','G'),('A','H'),('B','A'),('C','A'),('C','B'),('C','D'),
                  ('C','E'),('C','F'),('C','G'),('C','H'),('D','B'),('E','A'),('E','B'),('E','C'),('E','D'),('E','F'),('E','G'),
                  ('E','H'),('F','C'),('G','A'),('H','A'),('H','B'),('H','C'),('H','D'),('H','E'),('H','F'),('H','G')])
plt.figure(figsize =(9, 9))
nx.draw_networkx(G, node_color ='green')
```

```python
import numpy as np
B = nx.to_numpy_matrix(G)
for i in range(0, len(B)):
    for j in range(0, len(B)):
        print(B[i,j], end = " ")
    print()
```

```python
current_sum = 0
for i in range(0, len(B)):
    for j in range(0, len(B)):
        current_sum = current_sum + B[j,i]
    
    print("Person ", chr(i+65), " is the thief if ", int(current_sum), " people/person are telling the truth.")
    current_sum = 0
```
